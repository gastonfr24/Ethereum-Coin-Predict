<h1 style='color: green;'> Preprocesamiento de los datos </h1>
Quitaremos las variables que no sean necesesarias para el modelo y escalaremos las variables contínuas

```python
training_set = train_df.iloc[:,1:2].values
```

```python
training_set
```

```python
# Escalado de variables
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler()

training_set_scaled = sc.fit_transform(training_set)
```

```python
print(training_set_scaled.max())
print(training_set_scaled.min())
print(training_set_scaled)
```

```python
X_train = []
y_train = []

for i in range(60,len(training_set)):
    
    X_train.append(training_set_scaled[i-60:i, 0])
    y_train.append(training_set_scaled[i, 0])
    
X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
X_train.shape
```

```python
X_train[:,0]
```

```python
# Redimensión del set de train
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
X_train.shape
```

```python
X_train[:,:,0]
```

<h1 style='color:green;'>Creación de Red Neuronal Recurrente</h1>

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
```

```python
# Arquitectura del modelo
# Inicialización
model = Sequential()
```

```python
# Capas de input
model.add(LSTM(64, return_sequences = True, input_shape = (X_train.shape[1],1)))
model.add(Dropout(0.25))
#375
```

```python
# Capas ocultas
model.add(LSTM(64, return_sequences = True))
model.add(Dropout(0.25))

model.add(LSTM(64, return_sequences = True))
model.add(Dropout(0.25))


model.add(LSTM(64, return_sequences = False))
model.add(Dropout(0.25))
```

```python
# Capa de output
model.add(Dense(1))
```

```python
#model.summary()
```

```python
# Compilación de RNR
model.compile(optimizer='Nadam', loss='mean_squared_error')
```

```python
# Entrenamiento del modelo
model.fit(X_train, y_train, epochs=50, batch_size=32)
```

```python
model.save_weights('another_model.h5')
```

<h1 style='color:green;'>Testeo del Modelo</h1>

```python
# Testeo del modelo
test_stock_df = pd.read_csv('eth-usd-test.csv')
# Usaremos el precio de Abertura
real_stock_df = test_stock_df.iloc[:,1:2].values
real_stock_df.shape
```

```python
real_stock_df
```

```python
# Creamos un dataframe con los de train y test
all_df = pd.concat((train_df['price'], test_stock_df['price']), axis=0)
```

```python
inputs = all_df[len(all_df)-len(test_stock_df)-60:].values
inputs = inputs.reshape(-1, 1)
```

```python
inputs.shape
```

```python
# escalado de los ultimos días
inputs = sc.transform(inputs)
```

```python
# Creación de estructura con 60 timesteps y 1 salida.
X_test = []

for i in range(60, len(inputs)):
    X_test.append(inputs[i-60:i, 0])
       
X_test= np.array(X_test)
```

```python
# Redimensión de datos
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

```python
# Haciendo predicción
y_pred = sc.inverse_transform(model.predict(X_test))
```

```python
# Visualización de los resultados
plt.plot(real_stock_df, color='g', label='Precio Real de ETH')
plt.plot(y_pred, color='b', label='Precio Predicho de ETH')
plt.title('Predicción de una RNR del valor de ETH')
plt.legend()
plt.show()
```

<h1 style='color:green;'>Mejorar Red Neuronal</h1>

```python
def model_regressor(optimizer):   
    # Inicialización
    model = Sequential()

    # Capas de input
    model.add(LSTM(64, return_sequences = True, input_shape = (X_train.shape[1],1)))
    model.add(Dropout(0.25))

    # Capas ocultas
    model.add(LSTM(64, return_sequences = True))
    model.add(Dropout(0.25))
    model.add(LSTM(64, return_sequences = True))
    model.add(Dropout(0.25))
    model.add(LSTM(64, return_sequences = False))
    model.add(Dropout(0.25))

    # Capa de output
    model.add(Dense(1))

    # Compilación de RNR
    model.compile(optimizer=optimizer, loss='mean_squared_error')
    
    return model
```

```python
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasRegressor
#from scikeras.wrappers import KerasRegressor

regressor = KerasRegressor(build_fn = model_regressor)

parameters = {
    'batch_size' : [32],
    'epochs' : [25,50], 
    'optimizer' : ['Adam', 'Nadam', 'RMSprop']
}

grid_search = GridSearchCV(estimator = regressor, 
                           param_grid = parameters, 
                           scoring = 'neg_mean_squared_error', 
                           cv = 10
                          )
#grid_search = grid_search.fit(X_train, y_train)
```
